# Implementation

```python
import csv
import textwrap
import re
```

```python
def read_verse_data(csv_file):
    verse_data = {}
    with open(csv_file, 'r', encoding='utf-8') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            verse_id = row['id'].lower()
            verse_data[verse_id] = {
                'ChapterName': row['Chapter Name'],
                'Verse': row['Verse'],
                'Translation': row['Translation'],
                'Explanation': row['Explanation']
            }
    return verse_data
```

```python
def clean_and_wrap(text):
    cleaned_text = re.sub(r'\s+', ' ', text.strip())
    return textwrap.fill(cleaned_text, width=40)
```

```python
def print_aligned(title, text):
    print(f"{title}:".ljust(15))
    for line in textwrap.wrap(text, width=40):
        print(line)
```

```python
csv_file_path = '/content/new_data_thirukural'
verse_data = read_verse_data(csv_file_path)
```

# Results

```python
while True:
    user_input = input("User: ")
    if user_input.lower() == 'exit':
        print("Chatbot: Exiting. Goodbye!")
        break

    if user_input.lower() in verse_data:
        verse_info = verse_data[user_input.lower()]

        print("Chatbot:")
        print_aligned("Chapter Name", verse_info['ChapterName'])
        if 'Verse' in verse_info:
            print_aligned("Verse", clean_and_wrap(verse_info['Verse']))
        if 'Translation' in verse_info:
            print_aligned("Translation", clean_and_wrap(verse_info['Translation']))
        if 'Explanation' in verse_info:
            print_aligned("Explanation", clean_and_wrap(verse_info['Explanation']))
    else:
        print("Chatbot: I'm sorry, the provided ID was not found in the CSV file.")
```
